Data Exploration and Cleaning

```python
# Load a dataset and display the first five rows.
data=pd.read_csv("Demography_of_India.csv")
```

```python
data.head(5)
```

```python
# Display the basic information (column names, data types, etc.) of the dataset.
data.info()
```

```python
# Count the number of missing values in each column.
data.isnull().sum()
```

```python
# Rename the column District_code to District_ID.
data.rename(columns={"District_code":"District_ID"})
```

```python
# Create a new column Total_Population as the sum of Male and Female.
data["Total_Population"]=data["Male"]+data["Female"]
data.head(1)
```

```python
# Convert the State_name column to lowercase.
data["State_name"]=data["State_name"].str.lower()
data.head(1)
```

```python
# Check for duplicate rows and remove them if found.
data.isnull().sum()
```

```python
# Calculate the total population for each district.
data.groupby("District_name")["Age_Group_0_29"].mean()
```

```python
# What is the median value of Male_Workers in the dataset?
data["Male_Workers"].median()
```

```python
# Calculate the percentage of literate people in each district based on the Population column.
data["percent_of_Literate"]=(data["Literate"]/data["Total_Population"])*100;
data.groupby("District_name")["percent_of_Literate"].sum().round()
```

```python
# Find the district with the highest number of Female_Workers.
data.loc[data["Female_Workers"].idxmax(), "District_name"]
```
